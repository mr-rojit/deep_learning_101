```python
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim
import torch.functional as F
from torch.utils.data import DataLoader, TensorDataset

from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, roc_curve, precision_recall_curve, auc

# local import
# from base import BaseModel
```

```python
# generating data
X, y = make_moons(n_samples=200, noise=0.1, random_state=69)
X.shape, y.shape
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
# splitting data to train and validation set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# converting ndarray to tensor and building dataset and dataloaders
X_train = torch.as_tensor(X_train)
X_val = torch.as_tensor(X_test)
y_train = torch.as_tensor(y_train)
y_val = torch.as_tensor(y_test)


# dataset
train_dataset = TensorDataset(X_train, y_train)
val_dataset = TensorDataset(X_val, y_val)


# loader
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16)

len(train_dataset), len(val_dataset)
```
