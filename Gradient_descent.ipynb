```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# NN with 1 input and 1 output
```

```python
inp = 0.5
w = 0.5
goal = 0.8
weights_list = []
error_list = []
def neural_network(inp, weight):
    return inp * weight

for i in range(10):
    pred = neural_network(inp, w)
    delta =  pred - goal
    error = delta ** 2
    weight_delta = inp * delta
    w = w -  weight_delta
    weights_list.append(w)
    error_list.append(error)
    
```

```python
plt.scatter(weights_list, error_list, alpha=0.3)
```

```python
print(weights_list)
print()
print(error_list)
```

Here we can see the error is approaching 0. lets try with diffrent sets  of numbers

```python
inp = 2
w = 1
goal = 4
weights_list = []
error_list = []
def neural_network(inp, weight):
    return inp * weight

for i in range(6):
    pred = neural_network(inp, w)
    delta =  pred - goal
    error = delta ** 2
    weight_delta = inp * delta
    w = w -  weight_delta
    weights_list.append(w)
    error_list.append(error)
plt.scatter(weights_list, error_list, alpha=0.3)  
```

## Using alpha value

```python
alpha = 0.1
inp = 2
w = 1
goal = 4
weights_list = []
error_list = []
def neural_network(inp, weight):
    return inp * weight

for i in range(6):
    pred = neural_network(inp, w)
    delta =  pred - goal
    error = delta ** 2
    weight_delta = inp * delta
    w = w -  weight_delta * alpha
    weights_list.append(w)
    error_list.append(error)
plt.scatter(weights_list, error_list, alpha=0.3) 
```

```python
print(weights_list)
print()
print(error_list)
```

Now the weights are updated with low value and error is approaching zero
