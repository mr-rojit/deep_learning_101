#### Here, we are going to go through simple derivative; both single varibale and multivariable

```python
# importing basic libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# These are our functions of which we will calculate derivative of
def poly_func(x):
    # 2x^2 - 3x + 5
    return 2*x**2 - 3*x + 5

def squared(x):
    return x**2
```

```python
# plotting squared function
sq_x = np.arange(-4, 5, 1)
sq_y = squared(sq_x)
plt.plot(sq_x, sq_y)
```

```python
# plotting poly_func 
ploy_x = np.arange(-4, 5, 1)
poly_y = poly_func(sq_x)
plt.plot(ploy_x, poly_y)
```

```python
def scalter_derivative(function, x):
    h = 0.001
    return (function(x+h)- function(x)) / h
```

```python
scalter_derivative(squared,5)
```

```python
scalter_derivative(poly_func, 2)
```
