```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate data
n = 10
actual_w = 3
actual_b = 1
```

```python
# adding error to our data
np.random.seed(42)
X = np.random.rand(n,1)
e = np.random.rand(n,1) * 0.7
y = actual_w * X + actual_b + e
plt.scatter(X,y)
```

```python
print(X)
print('--**--')
print(y)
```

```python
np.random.seed(42)
w = np.random.randn(1)
b = np.random.randn(1)
w, b
```

```python
y_pred = w * X + b
plt.scatter(X, y)
plt.scatter(X, y_pred)
```

```python
np.random.seed(42)

w = np.random.randn(1)
b = np.random.randn(1)

lr = 0.1
epochs = 100
losses = []
ws = []
bs = []
for epoch in range(epochs):
    y_pred = w * X + b
    delta = y- y_pred

    loss = (delta ** 2).mean()
    ws.append(w)
    bs.append(b)
    losses.append(loss)

    b_grad = -2 * delta.mean()
    w_grad = -2 * (X * delta).mean()

    b -= lr * b_grad
    w -= lr * w_grad
print(w, b)
```

```python
new_pred = w * X + b
```

```python
plt.scatter(X, y)
plt.scatter(X, new_pred)
```
